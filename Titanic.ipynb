```python
df["Embarked"] = df["Embarked"].astype(str)
boarding_point = pd.get_dummies(df["Embarked"], drop_first=True,dtype=int)
```

```python
df.drop(["Name","Ticket","PassengerId","Embarked"],axis=1,inplace=True)
```

```python
x=df.drop("Survived",axis=1)
```

```python
x
```

```python
y=df["Survived"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LogisticRegression(max_iter=500)
```

```python
model.fit(x_train, y_train)
```

```python
predicted_res=model.predict(x_test)
predicted_res
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,predicted_res)
```

```python
confusion_matrix(y_test,predicted_res)
```

```python
90->1   survived  : model->survived
15->0   not_survived : model->not survived
20->0   not_survived : model->survived
54->1   survived  : model->not survived
```

```python
print(classification_report(y_test,predicted_res))
```
